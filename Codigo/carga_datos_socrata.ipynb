```python
import pandas as pd 
from sodapy import Socrata
import os
from dotenv import load_dotenv

load_dotenv('.env')

TOKEN_APP = os.getenv("TOKEN_SODAPY")
socrata_dataset_identifier = os.getenv("DATASET_ID")

client = Socrata("www.datos.gov.co", 
                 TOKEN_APP)
```

```python
socrata_dataset_identifier = 'jbjy-vk9h'

Query = """
select 
    id_contrato, nombre_entidad, departamento, descripcion_del_proceso, valor_del_contrato
where
    fecha_de_firma >= '2024-01-01'
limit
10000000
"""
results = client.get(socrata_dataset_identifier, query=Query)
```

```python
results_df = pd.DataFrame.from_records(results)
```
